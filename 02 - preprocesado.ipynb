Cargar el archivo train.csv

```python
TR = pd.read_csv("train.csv")
```

```python
TR.head(10)
```

Limpiamos columnas duplicadas

```python
if "ID" in TR.columns:
    TR = TR.drop(columns=["ID"])
TR = TR.loc[:, ~TR.columns.duplicated()]

print("Columnas actuales:", len(TR.columns))
display(TR.head(2))
```

Imputacion de valores

```python
TR.isnull().sum()
```

```python
num_cols = TR.select_dtypes(include=["int64", "float64"]).columns
imputer_num = SimpleImputer(strategy="mean")
TR[num_cols] = imputer_num.fit_transform(TR[num_cols])

# se llenan datos con los valores mas freceuntes, por decir moda
categoria_cols = TR.select_dtypes(include=["object"]).columns
imputer_cat = SimpleImputer(strategy="most_frequent")
TR[categoria_cols] = imputer_cat.fit_transform(TR[categoria_cols])


print("Valores nulos restantes:", TR.isnull().sum().sum())
display(TR.head(10))
```

```python
display(TR[categoria_cols].head(3))

TR_encoded = pd.get_dummies(TR, columns=categoria_cols, drop_first=True)

print("One-Hot ")
print("Antes:", len(categoria_cols), "columnas categóricas - Después:", TR_encoded.shape[1], "columnas totales.")
display(TR_encoded.head(3))
```

```python
scaler = MinMaxScaler()
TR_encoded[num_cols] = scaler.fit_transform(TR_encoded[num_cols])

print("Normalización completada entre 0 y 1")
display(TR_encoded[num_cols].head(3))
```
